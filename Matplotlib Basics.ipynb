**Line Graph**

```python
x = [1, 2, 3, 4, 5]
y = [3, 7, 4, 12, 8]
```

```python
plt.plot(x, y)

plt.title("Line graph")
plt.ylabel("y-axis")
plt.xlabel("x-axis")
plt.show
```

```python
plt.plot(x, y, '^')
plt.plot(x, y)

plt.title("Line graph")
plt.ylabel("y-axis")
plt.xlabel("x-axis")
plt.show
```

```python
x = np.arange(0, 3*np.pi, 0.1)
y = np.sin(x)

plt.plot(x, y, '^')
plt.plot(x, y)

plt.title('Sine Graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show
```

**Scatter Plot**

```python
x = [1, 2, 3, 4, 5]
y = [3, 7, 4, 12, 8]
```

```python
plt.scatter(x, y)

plt.title("Line graph")
plt.ylabel("y-axis")
plt.xlabel("x-axis")
plt.show
```

**Comparison**

```python
x = [1, 2, 3, 4, 5]
y = [3, 7, 4, 12, 8]

x2 = [5, 2, 7, 9, 3]
y2 = [2, 5, 7, 2, 4]
```

```python
plt.plot(x, y, color = 'g', linewidth = 6)
plt.plot(x2, y2, color = 'r', linewidth = 4)

plt.title('Comparison Graph')
plt.xlabel('x-axis')
plt.ylabel('y-label')

plt.annotate(xy = [2, 3], text = 'First Line')
plt.annotate(xy = [4, 4], text = 'Second Line')

plt.legend(['First_Line', 'Second_Line'], loc = 4)

plt.show
```

**Bar Graphs**

```python
#vertical bar graph

x = [1, 2, 3, 4, 5]
y = [3, 7, 4, 12, 8]

x2 = [1, 4, 6, 8, 13]
y2 = [2, 5, 7, 12, 23]

plt.bar(x, y, color = 'g', linewidth = 6)
plt.bar(x2, y2, color = 'r', linewidth = 4)

plt.title('Comparison Graph')
plt.xlabel('x-axis')
plt.ylabel('y-label')

plt.legend(['line1', 'line2'], loc = 4)

plt.show
```

```python
#horizontal bar graph

x = [1, 2, 3, 4, 5]
y = [3, 7, 4, 12, 8]

x2 = [1, 4, 6, 8, 13]
y2 = [2, 5, 7, 12, 23]

plt.barh(x, y, color = 'g', linewidth = 6)
plt.barh(x2, y2, color = 'r', linewidth = 4)

plt.title('Comparison Graph')
plt.xlabel('x-axis')
plt.ylabel('y-label')

plt.legend(['line1', 'line2'], loc = 4)

plt.show
```

**Bubble Graph and Pie Chart**

```python
#bubble graph - other variation of scatter plot
x = [1, 2, 3, 4, 5]
y = [3, 7, 4, 12, 8]

plt.scatter(x, y, s = np.random.rand(5) * 1000, c = np.random.rand(5))

plt.title("Line graph")
plt.ylabel("y-axis")
plt.xlabel("x-axis")
plt.show
```

```python
data = [3, 7, 4, 12, 5]
ex = (0, 0.1, 0, 0, 0)
lb = ['first', 'second', 'third', 'fourth', 'fifth']
plt.pie(data, labels = lb, explode = ex, shadow = True)
plt.show
```

```python
np.random.seed(42)
family_ivan = np.random.poisson(5, 5)
family_sid = np.random.poisson(5, 5)

x = np.arange(0, 10, 2)
plt.scatter(x, family_ivan, label='Ивановы')
plt.scatter(x, family_sid, label='Сидоровы')
plt.title('Статистика больных')
plt.ylabel('Кол-во больных')
plt.xlabel('T температура')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot(x, family_ivan, '--', label='Ивановы')
plt.plot(x, family_sid, '--', label='Сидоровы')
plt.title('Статистика больных')
plt.ylabel('Кол-во больных')
plt.xlabel('T температура')
plt.legend()
plt.show()
```

```python
# сгенерируем 200 случайных точек
import scipy.stats as sps

sample = sps.norm.rvs(size=200)
```

```python
plt.figure(figsize=(10, 1))
plt.scatter(sample, np.zeros(200), alpha=0.2)  # alpha - прозрачность точки
plt.show()
```

```python
grid = np.linspace(-3, 3, 100)

plt.figure(figsize=(10, 4))
plt.scatter(sample, np.zeros(200) - 0.02, alpha=0.2, label='sample')
plt.plot(grid, sps.norm.pdf(grid), color='red', label='density')
plt.legend()
plt.show()
```
