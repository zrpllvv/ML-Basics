**Dealing with missing values**

```python
#null values
data = {
    'place' : [1, 2, 3, 4],
    'games' : [32, 33, 33, 32],
    'points' : [np.nan,73, 72, 72]
}
```

```python
df1 = pd.DataFrame(data)
df1
```

```python
#isnull function
df1.isnull()
```

```python
df1.isnull().sum()
```

```python
df2 = df1.fillna(1)
#filling null values with the arguement in the functions
```

```python
df2
```

```python
#dropping Null values:deleting entire row.
a = df1.dropna()
a
```

```python
#dropping row with null values
a = df1.dropna(axis = 0)
a
```

```python
#dropping column with null values
a = df1.dropna(axis = 1)
a
```

```python
#creating a data with non NULL values
a = pd.notnull(df1['points'])
a
```

```python
df1[a]
#removed false entries.
```
